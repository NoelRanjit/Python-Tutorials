```python
import pandas as pd

people = {
    'first' : ['Corey', 'Jane', 'John', 'Adam', 'Morty'],
    'last' : ['Schafer', 'Doe', 'Doe', 'Doe', 'Smith'],
    'email' : ['cschafer@gmail.com', 'janedoe@gmail.com', 'johndoe@gmail.com', 'adoe@gmail.com', 'ms@gmail.com']
}

df = pd.DataFrame(people)
df
```

```python
# sorting values by last name. no particular order. only default
df.sort_values(by = 'last')
```

```python
# having first name ascending and last name descending
df.sort_values(by = ['last', 'first'], ascending = [False,True], inplace = True)
df
# it gives priority to whatever was placed first. if ['last', 'first'], last is prioritised
# this arranges the items in last by descending order first and then if they have the same last name, sorts the 
# first name by ascending order
```

```python
# simply arranges the first name by ascending order
df.sort_values(by = ['first', 'last'], ascending = ['True', 'False'])
```

```python
# the indexes are all over the place so we can sort them. 
# Note : it doesn't return the ascending/descending changes. It reverts back
df.sort_index()
# to be able to keep changes and rearrange the indexes: 
# df.sort_values(by = ['last', 'first'], ascending = [False,True], ignore_index = True)
# second alternative: df.reset_index(drop = True)
```

```python
# displaying only sorted last names
df['last'].sort_values()
```

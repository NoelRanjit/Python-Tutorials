```python
### Question: Find out what percentage of people from each country know python
```

```python
# First we find out the responses from each country
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
# then we find out how many know python from each country
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x : x.str.contains('Python').sum())
country_uses_python
```

```python
# using concat funtion create a df with relevant values
python_df = pd.concat([country_respondents, country_uses_python], axis = 'columns', sort = False)
python_df
# axis = 'columns' helps to concat as columns. otherwise they would concat into rows as default (not what we want)
```

```python
python_df['PythonPct'] = (python_df['LanguageWorkedWith']/python_df['Country']) * 100
python_df
```

```python
# arranging in descending order
python_df.sort_values(by ='PythonPct', ascending = False)
```

```python
# display all pokemon with Mega in their name

data.loc[data['Name'].str.contains("Mega")]
```

```python
# print all pokmeon names except mega pokemon
data.loc[~data['Name'].str.contains("Mega")]
```

```python
# use regular expressions to get pokemon that are type fire or grass
import re
data.loc[data["Type 1"].str.contains("Fire|Grass" , regex = True)]
```

```python
# using regular expressions display all pokemon with pi in their name
import re
data.loc[data["Name"].str.contains("pi[a-z]*", regex = True)]
```

```python
# using regular expressions display pokemon whose name start with pi
data.loc[data['Name'].str.contains('^pi[a-z]*' ,flags = re.I, regex = True)]
```

```python
# change pokemon type Fire to Flame
data.loc[data['Type 1'] == 'Fire', 'Type 1'] = 'Flame'
data
```

```python
# change all flame pokemon to legendary
data.loc[data["Type 1"] == 'Flame' , 'Legendary' ] = True
data
```

```python
# if its a first generation pokemon set generation to test 1 and legendary to test 2
data.loc[data['Generation'] == 1, ['Generation', 'Legendary']] = ['Test 1' , 'Test 2']
data
```

```python
# using groupby find the pokemon with the pokemon type1 with the highest defense
data.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)
```

```python
# using groupby find the pokemon with the pokemon type1 with the highest hp
data.groupby(['Type 1']).mean().sort_values('HP', ascending = False)
```

```python
# count all the type 1 pokemon
data['count'] = 1
data.groupby(['Type 1']).count()['count']
```

```python
# count the number of type 2 pokemon for each type 1 pokemon
data['count'] = 1
data.groupby(['Type 1' , 'Type 2']).count()['count']
```
